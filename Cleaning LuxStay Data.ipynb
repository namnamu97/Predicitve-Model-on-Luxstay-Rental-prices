```python
import pandas as pd
import datetime
```

```python
x=datetime.datetime.today().strftime('%d-%m-%Y') #today date 
file_name_raw = f'luxstay_hanoi_{x}_raw.csv'
df=pd.read_csv(file_name_raw)
```

```python
df
```

```python
 #get the district from address
df['District'] = df['Address'].apply(lambda x: x.split(',')[0])
    
    #getthe type of building
df['Type']=df['Type'].str.strip() #there are two \n in this field, this removes the first one
df['Type of Home'] = df['Type'].apply(lambda x: x.split('·')[0]) 
df['Type of Home']= df['Type of Home'].str.strip() #remove the last \n left
    
    #geting building name from the dataframe
building_name=[]
for i in range(len(df['Type'])):
    a=df['Type'][i].split('·')
    if len(a) == 2:
        name=a[1] 
    else:
        name='None'
    building_name.append(name)
df['Building Name']= building_name
```

```python
#Seperate Guest, Bedroom, Bathroom to show the offered numbers but there are still words in it
for n in range(len(df['Description'])):
    y=df['Description'][n].split('·')
    if len(y) != 3:
        df.drop([n],inplace=True)

df['Guest'] = df['Description'].apply(lambda x: x.split('·')[0])
df['Bedroom'] = df['Description'].apply(lambda x: x.split('·')[1])
df['Bathroom'] = df['Description'].apply(lambda x: x.split('·')[2])
#remove the words
df['Guest'] = df['Guest'].apply(lambda x: x.split()[0])
df['Bedroom'] = df['Bedroom'].apply(lambda x: x.split()[0])
df['Bathroom'] = df['Bathroom'].apply(lambda x: x.split()[0])
```

```python
#get the labelled price
df['Price'] = df['Price'].apply(lambda x: x.split('/')[0])
df['Price'] = df['Price'].str.replace('$','') #remove the dollar sign
```

```python
#drop unesscessary columns
df.drop(['Description','Address','Type'],axis=1,inplace=True)
    
#format the data type of some columns
df[['Guest','Bedroom','Bathroom']] = df[['Guest','Bedroom','Bathroom']].astype(int)
df['Price'] = df['Price'].astype(float)
```

```python
#store the clean data in csv file
file_name_clean = f'luxstay_hanoi_{x}_clean.csv'
df.to_csv(file_name_clean, index=False, encoding='utf-8')
```
