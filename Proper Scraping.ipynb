```python
#include packages
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import datetime
```

```python
#the work of scraping
base_url='https://www.luxstay.com/vietnam/ha-noi?page='
#list to store data
links=[]
types=[]
prices=[]
locations=[]
descs=[]
#go through each page
for n in range(1,136):
    url= f'{base_url}{n}'
    print('Processing {} ...'.format(url))
    driver= webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    time.sleep(8) #need to have enough time for the request to load all content
    content = driver.page_source
    soup = BeautifulSoup(content)
    #find data
    for a in soup.findAll(attrs={'class':'promo__content'}):
        link= 'https://www.luxstay.com/' + a.find('div',attrs={'class':'is-relative'}).find('a')['href']
        cat=a.find('div', attrs={'class':'promo__type bold'})
        price=a.find('div', attrs={'class':'promo__price mb--6'})
        location=a.find('div', attrs={'class':'promo__add mb--6 promo__add mb--6'})
        desc=a.find('div', attrs={'class':'promo__description my--6 text-lowercase'})
        links.append(link)
        types.append(cat.text)
        prices.append(price.text)
        locations.append(location.text)
        descs.append(desc.text)        
```

```python
#store in dataframe
df = pd.DataFrame({'Link':links,'Type':types,'Price':prices,'Address':locations,'Description': descs})
#export to csv file
x=datetime.datetime.today().strftime('%d-%m-%Y') #today date 
file_name_raw = f'luxstay_hanoi_{x}_raw.csv'
df.to_csv(file_name_raw, index=False, encoding='utf-8')
```
